```python
from sklearn.datasets import load_wine

# 加载 wine 数据集
wine = load_wine()

# 查看 wine 数据集的键值
print(wine.keys())

# 查看数据集中的数据
print(wine.data)

# 查看数据集的形状（行数和特征数）
print(wine.data.shape)  # 输出：(178, 13)

# 查看数据集的标签（target）
print(wine.target)

# 查看标签的形状
print(wine.target.shape)  # 输出：(178,)

# 查看标签的类别名称
print(wine.target_names)

# 输出数据集中所有特征的名称
print(wine.feature_names)
```

```python
# 1.导入 Pandas 包并转换数据集为 DataFrame
import pandas as pd

# 将 wine 数据集转换为 DataFrame
df_wine = pd.DataFrame(wine.data)

# 查看前 5 行数据
df_wine.head()
```

```python
# 2.设置 DataFrame 的列名
# 为 DataFrame 添加特征列名
df_wine.columns = wine.feature_names

# 再次查看前 5 行数据，验证设置是否成功
df_wine.head()
```

```python
# 3.将标签添加为新列
# 添加标签列 'wine_class'
df_wine['wine_class'] = wine.target

# 查看 DataFrame 的最后 5 行
df_wine.tail()
```

```python
# 4.删除标签列并将其与特征数据分离
# 删除标签列，保留特征数据
X = df_wine.drop('wine_class', axis=1)

# 将标签列单独赋值给 y
y = df_wine['wine_class']
```

```python
# 5.使用 scikit-learn 进行训练集和测试集的拆分
from sklearn.model_selection import train_test_split

# 拆分训练集和测试集，测试集占 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# 6.构建多层感知机模型 (MLP)
from sklearn.neural_network import MLPClassifier

# 创建 MLP 分类器，隐藏层有 500 个神经元，使用 lbfgs 作为优化器
#model = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(500,))
model = MLPClassifier(hidden_layer_sizes=(500,), solver='lbfgs', max_iter=1000)
```

```python
# 7. 训练模型
# 在训练集上训练模型
model.fit(X_train, y_train)
```

```python
# 8.在训练集和测试集上进行预测
# 在训练集上进行预测
y_predict_on_train = model.predict(X_train)

# 在测试集上进行预测
y_predict_on_test = model.predict(X_test)
```

```python
# 9.评估模型性能
from sklearn.metrics import accuracy_score

# 计算训练集和测试集的准确率
print("训练集的准确率为: {:.2f}%".format(100 * accuracy_score(y_train, y_predict_on_train)))
print("测试集的准确率为: {:.2f}%".format(100 * accuracy_score(y_test, y_predict_on_test)))
```

```python
# 10.数据标准化处理
from sklearn.preprocessing import StandardScaler

# 创建 StandardScaler 对象
scaler = StandardScaler()

# 在训练集上拟合并转换数据
X_train = scaler.fit_transform(X_train)

# 在测试集上进行转换
X_test = scaler.transform(X_test)
```

```python
# 11.使用标准化数据再次训练模型
# 使用相同的参数创建新的 MLP 分类器
model = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(100,))

# 训练模型
model.fit(X_train, y_train)

# 在训练集和测试集上进行预测
y_predict_on_train = model.predict(X_train)
y_predict_on_test = model.predict(X_test)

# 评估模型性能
print("训练集的准确率为: {:.2f}%".format(100 * accuracy_score(y_train, y_predict_on_train)))
print("测试集的准确率为: {:.2f}%".format(100 * accuracy_score(y_test, y_predict_on_test)))
```

```python
# 12.查看模型参数
# 查看模型的偏置参数和权重
print(model.coefs_[0])  # 第一层的参数
print(model.coefs_[1])  # 第二层的参数
```
